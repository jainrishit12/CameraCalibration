1. Importing all functions from calib.py file

```python
from calib import *
```

2. Displaying images before processing

```python
image_filenames = [ 'IMG (1).jpg',
                    'IMG (2).jpg',
                    'IMG (3).jpg',
                    'IMG (4).jpg',
                    'IMG (5).jpg']

for filename in image_filenames:
    image = load(filename)
    display(image)
```

3. Calibration

```python
ret, mtx, dist,rvecs,tvecs = calibrate()
print("Camera Matrix:")
print(mtx)
```

4. Undistortion

```python
count = 1
for filename in image_filenames:
    image = load(filename)
    undistortedIMG = undistort(image, mtx, dist)
    display(undistortedIMG)
    cv.imwrite('undistorted' + str(count) + '.jpg' , undistortedIMG)
    count+= 1
```
